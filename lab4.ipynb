```python
# Required inputs
import os
import requests
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile

print("Libs have been imported successfully!")
```

```python
# Still the same downloader...
def downloader(url):
    if not os.path.exists("data"):
        os.makedirs("data")
    response = requests.get(url)
    if response.status_code == 200:
        filename = os.path.join("data", url.split('/')[-1])
        with open(filename, "wb") as file:
            file.write(response.content)
        with zipfile.ZipFile(file, 'r') as zip:
            zip.extractall("data")
        print(f"File {filename} has been downloaded successfully!")
        return filename
    else:
        print(f"Error downloading file. Status code: {response.status_code}")
        return None

url = "https://archive.ics.uci.edu/static/public/235/individual+household+electric+power+consumption.zip"
# file = downloader(url)
file = "data/household_power_consumption.txt"
```

# №1 Pandas frame

```python
start = datetime.now()
df = pd.read_csv(file, sep=';', na_values='?')
df.dropna(inplace=True)
end = datetime.now()
print(f"Execution time is {end - start}")
print(df.dtypes)

df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
df.tail(10)
```

1. Обрати всі домогосподарства, у яких загальна активна споживана потужність перевищує 5 кВт.

```python
start = datetime.now()
power = df[df['Global_active_power'] > 5]
print(f"Rows: {len(power)}")
end = datetime.now()
print(f"Time: {end - start}")
power.head()
```

2. Обрати всі домогосподарства, у яких вольтаж перевищую 235 В.

```python
start = datetime.now()
voltage = df[df['Voltage'] > 235]
print(f"Rows: {len(voltage)}")
end = datetime.now()
print(f"Time: {end - start}")
voltage.head()
```

3. Обрати всі домогосподарства, у яких сила струму лежить в межах
19-20 А, для них виявити ті, у яких пральна машина та холодильних
споживають більше, ніж бойлер та кондиціонер.

```python
start = datetime.now()
curStrength = df[(df['Global_intensity'] >= 19) & (df['Global_intensity'] <= 20) & (df['Sub_metering_2'] > df['Sub_metering_3'])]
print(f"Rows: {len(curStrength)}")
end = datetime.now()
print(f"Time: {end - start}")
curStrength.head()
```

4. Обрати випадковим чином 500000 домогосподарств (без повторів
елементів вибірки), для них обчислити середні величини усіх 3-х
груп споживання електричної енергії, а також

```python
start = datetime.now()
householdsRandomized = df.sample(n=500000, replace=True)
avgSubMetering1 = householdsRandomized['Sub_metering_1'].mean()
avgSubMetering2 = householdsRandomized['Sub_metering_2'].mean()
avgSubMetering3 = householdsRandomized['Sub_metering_3'].mean()
end = datetime.now()
print(f"Time: {end - start}")
print("[Avg] sub_metering_1:", avgSubMetering1)
print("[Avg] sub_metering_2:", avgSubMetering2)
print("[Avg] sub_metering_3:", avgSubMetering3)
```

5. Обрати ті домогосподарства, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другоої полвини.

```python
start = datetime.now()
# households consuming > 6 kW after 6pm
filtered = df[(df['Time'] > pd.to_datetime('18:00:00').time()) & (df['Global_active_power'] > 6)]
filtered = filtered[(filtered['Sub_metering_2'] > filtered['Sub_metering_1']) & (filtered['Sub_metering_2'] > filtered['Sub_metering_3'])]
# choose each 3 result from first half and each 4 result from second one
half1 = filtered.iloc[:len(filtered) // 2].iloc[2::3]
half2 = filtered.iloc[len(filtered) // 2:].iloc[3::4]
concatanetion = pd.concat([half1, half2])
end = datetime.now()
print(f"Time: {end - start}")
print(f"DF rows: {concatanetion.shape[0]}")
concatanetion.head(5)
```

# №2 Numpy array

```python
start = datetime.now()
data = np.genfromtxt(file, delimiter=';', skip_header=1, dtype=str)
mask = np.all(data != '?', axis=1)
data = data[mask]
dataFloat = data[:, 2:].astype(float)
data = np.hstack([data[:, :2], dataFloat], dtype=object)
end = datetime.now()
print(f"Time: {end - start}")
print(data)
```

1. Обрати всі домогосподарства, у яких загальна активна споживана
потужність перевищує 5 кВт.

```python
start = datetime.now()
columnNumeric = data[:, 2].astype(float)
indices = np.where(columnNumeric > 5)
NPpower = data[indices]
print(f"Rows: {len(NPpower)}")
end = datetime.now()
print(f"Time: {end - start}")
print(NPpower[:5])
```

2. Обрати всі домогосподарства, у яких вольтаж перевищую 235 В.

```python
start = datetime.now()
columnNumeric = data[:, 4].astype(float)
indices = np.where(columnNumeric > 235)
NPvoltage = data[indices]
print(f"Rows: {len(NPvoltage)}")
end = datetime.now()
print(f"Time: {end - start}")
print(NPvoltage[:5])
```

3. Обрати всі домогосподарства, у яких сила струму лежить в межах
19-20 А, для них виявити ті, у яких пральна машина та холодильних
споживають більше, ніж бойлер та кондиціонер.

```python
start = datetime.now()
curStrength = data[(data[:,5] >= float(19)) & (data[:,5] <= float(20)) & (data[:,7] > data[:,8])]
print(f"Rows: {len(curStrength)}")
end = datetime.now()
print(f"Time: {end - start}")
print(curStrength[:5])
```

4. Обрати випадковим чином 500000 домогосподарств (без повторів
елементів вибірки), для них обчислити середні величини усіх 3-х
груп споживання електричної енергії, а також

```python
randomInd = np.random.choice(data.shape[0], replace=True, size=500000)
householdsRandomized = data[randomInd]
sub_metering_1_values = householdsRandomized[:, 6]  
sub_metering_2_values = householdsRandomized[:, 7]  
sub_metering_3_values = householdsRandomized[:, 8]  
avgSubMetering1 = np.mean(sub_metering_1_values)
avgSubMetering2 = np.mean(sub_metering_2_values)
avgSubMetering3 = np.mean(sub_metering_3_values)
end = datetime.now()
print(f"Time: {end - start}")
print("[Avg] sub_metering_1:", avgSubMetering1)
print("[Avg] sub_metering_2:", avgSubMetering2)
print("[Avg] sub_metering_3:", avgSubMetering3)
```

5. Обрати ті домогосподарства, які після 18-00 споживають понад 6
кВт за хвилину в середньому, серед відібраних визначити ті, у яких
основне споживання електроенергії у вказаний проміжок часу
припадає на пральну машину, сушарку, холодильник та освітлення
(група 2 є найбільшою), а потім обрати кожен третій результат із
першої половини та кожен четвертий результат із другої половини.

```python
start = datetime.now()
# households consuming > 6 kW after 6pm
filtered = data[(data[:, 1] > '18:00:00') & (data[:, 2] > 6) & (data[:, 7] > data[:, 6]) & (data[:, 7] > data[:, 8])]
group1 = filtered[:len(filtered)//2]
group2 = filtered[len(filtered)//2:]
# choose each 3 result from first half and each 4 result from second one
each3 = group1[::3]
each4 = group2[::4]
concatanetion = np.concatenate((each3, each4), axis=0)
end = datetime.now()
print(f"Time: {end - start}")
print(f"DF rows: {concatanetion.shape[0]}")
print(concatanetion[:5])
```
